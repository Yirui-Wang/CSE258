```python
from collections import defaultdict
import pandas as pd
from matplotlib import pyplot as plt

df = pd.read_csv('data/fit.csv')
fit_count = defaultdict(int)
for fit in df['fit']:
    fit_count[fit] += 1
print(fit_count)

labels = []
sizes = []
for key, value in fit_count.items():
    labels.append(key)
    sizes.append(value)

plt.figure(figsize=(5,5))
colors = ['orange','green','lightblue']
patches,text1,text2 = plt.pie(sizes,
                              labels=labels,
                              colors=colors,
                              autopct = '%3.2f%%',
                              shadow = False,
                              startangle =90,
                              pctdistance = 0.6)
plt.axis('equal')
plt.title('Fit Distribution')
plt.savefig('fit.png')
plt.show()
```

```python
count = defaultdict(int)
for weight in df['weight']:
    key = (weight // 10) * 10
    count[key] += 1
print(count)
name_list = []
num_list = []
for key, value in count.items():
    name_list.append(key)
name_list = sorted(name_list)
print(name_list)
for name in name_list:
    num_list.append(count[name])
print(num_list)
plt.figure(figsize=(10, 5))
plt.bar(range(len(num_list)), num_list, tick_label=name_list)
plt.xlabel('Weight (lbs)')
plt.ylabel('Number')
plt.title('Weight Distribution')
plt.show()
```
