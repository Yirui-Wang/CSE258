```python
from collections import defaultdict
import pandas as pd
from matplotlib import pyplot as plt

df = pd.read_csv('data/fit.csv')
fit_count = defaultdict(int)
for fit in df['fit']:
    fit_count[fit] += 1
print(fit_count)

labels = []
sizes = []
for key, value in fit_count.items():
    labels.append(key)
    sizes.append(value)

plt.figure(figsize=(5,5))
colors = ['lightblue','lightgreen','yellow']
patches,text1,text2 = plt.pie(sizes,
                              labels=labels,
                              colors=colors,
                              autopct = '%3.2f%%',
                              shadow = False,
                              startangle =90,
                              pctdistance = 0.6)
plt.axis('equal')
plt.title('Fit Distribution')
plt.savefig('fit.png')
plt.show()
```

```python
count = defaultdict(int)
for weight in df['weight']:
    key = (weight // 10) * 10
    count[key] += 1
print(count)
name_list = []
num_list = []
for key, value in count.items():
    name_list.append(key)
name_list = sorted(name_list)
print(name_list)
for name in name_list:
    num_list.append(count[name])
print(num_list)
plt.figure(figsize=(10, 5))
plt.bar(range(len(num_list)), num_list, tick_label=name_list)
plt.xlabel('Weight (lbs)')
plt.ylabel('Number')
plt.title('Weight Distribution')
plt.show()
```

```python
count = defaultdict(int)
for cat in df['category']:
    count[cat] += 1
print(count)

labels = []
sizes = []
other = 0
for key, value in count.items():
    if value < 1000:
        other += value
    else:
        labels.append(key)
        sizes.append(value)
labels.append('other')
sizes.append(other)

plt.figure(figsize=(10,11))
colors = ['#ff6666', '#ffcc99', '#a3e861', '#66b3ff', '#c2c2f0', '#ffb3e6', '#cf9c5b']
patches,text1,text2 = plt.pie(sizes,
                              labels=labels,
                              colors=colors,
                              autopct = '%3.2f%%',
                              shadow = False,
                              startangle =90,
                              pctdistance = 0.8)
plt.axis('equal')
plt.title('Cloth Category Distribution')
plt.savefig('cloth_cat.png')
plt.show()
```
